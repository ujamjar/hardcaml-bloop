The core routines for drawing nice covers over the cubes.

```ocaml
let pi = 4. *. atan 1.
```

```ocaml
let outline_cubes = `O { P.o with P.width = 0.01; (*P.dashes=Some (0.113,[0.05;0.1])*) }
let outline_col = Color.(v 0.7 0.2 0.2 1.)
let shaded_col = Color.(v 0.7 0.2 0.2 0.1)

(* much pretty rendering of the covers *)
let pc0 r = P.empty >> P.circle P2.(v 0. 0.) r

let pc1 r = 
    P.empty >> P.sub P2.(v (-0.5) r) >> 
    P.line P2.(v 0. r) >> 
    P.earc ~cw:true Size2.(v r r) P2.(v 0. (-.r)) >> 
    P.line P2.(v (-0.5) (-.r))

let pc2 r = 
    P.empty >> P.sub P2.(v (-0.5) r) >> 
    P.line P2.(v 0. r) >> 
    P.earc ~cw:true Size2.(v r r) P2.(v r 0.) >> 
    P.line P2.(v r (-0.5))

let pc2c r = pc2 r >> P.line P2.(v (-0.5) (-0.5))

let img ?(area=`Anz) col p = I.cut ~area p (I.const col) 

let shaded0 r = 
    I.blend 
        (img ~area:outline_cubes outline_col (pc0 r)) 
        (img shaded_col (pc0 r))

let shaded1 r = 
    I.blend 
        (img ~area:outline_cubes outline_col (pc1 r)) 
        (img shaded_col (pc1 r))

let shaded2 r = 
    I.blend 
        (img ~area:outline_cubes outline_col (pc2 r)) 
        (img shaded_col (pc2c r))

let shaded3 r =
    I.blend
        (img ~area:outline_cubes outline_col (P.empty >> P.sub P2.(v (-. r) 0.5) >> P.line P2.(v (-. r) (-0.5))))
        (img shaded_col (P.empty >> P.rect Box2.(v P2.(v (-. r) (-0.5)) Size2.(v (1. -. (0.5 -. r)) 1.))))

let shaded4 r =
    I.blend
        (img ~area:outline_cubes outline_col 
            (P.empty >> P.sub P2.(v (-. r) 0.5) >> P.line P2.(v (-. r) (-0.5))
                     >> P.sub P2.(v r 0.5) >> P.line P2.(v r (-0.5))))
        (img shaded_col (P.empty >> P.rect Box2.(v P2.(v (-. r) (-0.5)) Size2.(v (r +. r) 1.))))

let shaded5 = img shaded_col (P.empty >> P.rect Box2.(v P2.(v (-0.5) (-0.5)) Size2.(v 1. 1.)))

let show img = Size2.v 20. 20., Box2.v (P2.v (-0.6) (-0.6)) (Size2.v 1.2 1.2), 
    (I.blend (I.move P2.(v (-0.5) (-0.5)) box1) img)

let cube_cover rad t l b r = 
    match t, l, b, r with
    (* no edges *)
    | false, false, false, false -> shaded5
    (* 1 edge *)
    | true , false, false, false -> shaded3 rad >> I.rot (3. *. pi /. 2.)
    | false, true , false, false -> shaded3 rad 
    | false, false, true , false -> shaded3 rad >> I.rot (pi /. 2.)
    | false, false, false, true  -> shaded3 rad >> I.rot pi
    (* 2 edges; l+r, t+b *)
    | true , false, true , false -> shaded4 rad >> I.rot (pi /. 2.)
    | false, true , false, true  -> shaded4 rad
    (* 2 edges; corners *)
    | true , true , false, false -> shaded2 rad >> I.rot (pi /. 2.)
    | true , false, false, true  -> shaded2 rad
    | false, true , true , false -> shaded2 rad >> I.rot pi
    | false, false, true , true  -> shaded2 rad >> I.rot (3. *. pi /. 2.)
    (* 3 edges *)
    | false, true , true , true  -> shaded1 rad >> I.rot (3. *. pi /. 2.)
    | true , false, true , true  -> shaded1 rad
    | true , true , false, true  -> shaded1 rad >> I.rot (pi /. 2.)
    | true , true , true , false -> shaded1 rad >> I.rot pi
    (* 4 edges *)
    | true , true , true , true  -> shaded0 rad

(*let _ = show @@ (shaded0 0.45)
let _ = show @@ I.blend (shaded1 0.45) (shaded1 0.43)
let _ = show @@ I.blend (shaded2 0.45) ((shaded2 0.43) >> I.rot (2. *. pi /. 2.))
let _ = show @@ shaded3 0.3
let _ = show @@ shaded4*)

let t, f, rad = true, false, 0.45
let _ = show @@ cube_cover rad f f f f
let _ = show @@ cube_cover rad t f f f
let _ = show @@ cube_cover rad f t f f
let _ = show @@ cube_cover rad f f t f
let _ = show @@ cube_cover rad f f f t
let _ = show @@ cube_cover rad f t f t
let _ = show @@ cube_cover rad t f t f
let _ = show @@ cube_cover rad t t f f
let _ = show @@ cube_cover rad t f f t
let _ = show @@ cube_cover rad f t t f
let _ = show @@ cube_cover rad f f t t
let _ = show @@ cube_cover rad f t t t
let _ = show @@ cube_cover rad t f t t
let _ = show @@ cube_cover rad t t f t
let _ = show @@ cube_cover rad t t t f
let _ = show @@ cube_cover rad t t t t

```

Piece things together and draw a full cover

```ocaml
let draw_cube rad l t w h = 
    let rec loop_y y = 
        if y = t+h then I.void
        else
            let t = y=t and b = y=(t+h-1) in
            let rec loop_x x = 
                if x = l+w then loop_y (y+1)
                else
                    let l = x=l and r = x=(l+w-1) in
                    let x' = float_of_int (x mod 4) and y' = float_of_int (3 - (y mod 4)) in
                    cube_cover rad t l b r >> I.move P2.(v (0.5 +. x') (0.5 +. y')) >> I.blend (loop_x (x+1))
            in
            loop_x l
    in
    loop_y t
```

Lets say we have a product from the set {1,2,3,4,-1,-2,-3,-4}.  There are no more than 4 terms, and no +ve and -ve terms (ie a.a' = 0).

1 term = 4x2 or 2x4, 2 terms = 2x2 or 4x1 or 1x4, 3 terms = 2x1 or 1x2, 4 terms = 1x1.

The following looks at the variable in a cover and works out the rect size and top left indices.  It's a wee but brute force, but I am not sure how to simplify it.

```ocaml
let select_column terms = 
    let contains = List.fold_left (fun a x -> a && List.mem x terms) true in
    (* ab *)
    if contains [-1;-2;] then 0
    else if contains [-1;2] then 1
    else if contains [1;2] then 2
    else if contains [1;-2] then 3
    else if contains [1] then 2
    else if contains [-1] then 0
    else if contains [2] then 1
    else if contains [-2] then 3
    else 0

let count_row terms = 
    let row_terms = [3;4;-3;-4] in
    List.fold_left (fun a x -> a + (if List.mem x row_terms then 1 else 0)) 0 terms

let count_col terms = 
    let col_terms = [1;2;-1;-2] in
    List.fold_left (fun a x -> a + (if List.mem x col_terms then 1 else 0)) 0 terms

let select_row terms = 
    let contains = List.fold_left (fun a x -> a && List.mem x terms) true in
    (* cd *)
    if contains [-3;-4;] then 0
    else if contains [-3;4] then 1
    else if contains [3;4] then 2
    else if contains [3;-4] then 3
    else if contains [3] then 2
    else if contains [-3] then 0
    else if contains [4] then 1
    else if contains [-4] then 3
    else 0

let select_cube terms = 
    let row, col = select_row terms, select_column terms in
    let crow, ccol = count_row terms, count_col terms in
    (*Printf.printf "%i %i %i %i\n" row col crow ccol;*)
    match List.length terms, crow, ccol with
    | 4, 2, 2 -> row, col, 1, 1
    | 3, 2, 1 -> row, col, 2, 1
    | 3, 1, 2 -> row, col, 1, 2
    | 2, 1, 1 -> row, col, 2, 2
    | 2, 2, 0 -> row, col, 4, 1
    | 2, 0, 2 -> row, col, 1, 4
    | 1, 1, 0 -> row, col, 4, 2
    | 1, 0, 1 -> row, col, 2, 4
    | 0, _, _ -> 0, 0, 4, 4
    | _ -> failwith "bad cube terms?"
```

```ocaml
let draw_kmap terms = 
    let img = List.fold_left (fun img terms ->
        let r,c,w,h = select_cube terms in
        I.blend (draw_cube 0.45 c r w h) img) I.void terms
    in
    let img = I.move P2.(v 0.75 0.25) (img >> I.blend (boxes d) >> I.blend labels) in
    img
```

```ocaml
let draw_kmaps maps = 
    let draw_row maps = 
        let _,img = Array.fold_left (fun (i,img) terms ->
            let kmap = I.move P2.(v (float_of_int i *. 5.) 0.) (draw_kmap terms) in
            i+1, I.blend img kmap) (0,I.void) maps
        in 
        img
    in
    let draw_rows maps = 
        let _, img = Array.fold_left (fun (i,img) maps ->
            let kmaps = I.move (P2.v 0. (float_of_int i *. -5.)) (draw_row maps) in
            i+1, I.blend img kmaps) (1,I.void) maps
        in
        img
    in
    let length = Array.length maps in
    let img = draw_rows maps in
    I.move (P2.v 0. (float_of_int length *. 5.)) img
```

```ocaml
let draw_kmap' t = Size2.v 80. 80., Box2.v (P2.v (0.) (0.)) (Size2.v 5. 5.), draw_kmap t
let draw_kmaps' t = 
    let rows = float_of_int @@ Array.length t in
    let cols = float_of_int @@ Array.fold_left (fun x a -> max x (Array.length a)) 0 t in
    Size2.v (cols *. 40.) (rows *. 40.), Box2.v (P2.v (0.) (0.)) (Size2.v (cols *. 5.) (rows *. 5.)), draw_kmaps t
```
